```python
import re
import requests
from datetime import datetime
r = requests.get('http://www.ppomppu.co.kr/search_bbs.php?bbs_cate=2&keyword=pass%BE%DB')
```

```python
#posts[0]
lastime = datetime.strptime("2020-07-13 12:11", '%Y-%m-%d %H:%M')
for post in posts:
    try:
        contents = re.findall("<a href=zboard.*?>(.*?)</a>", post, re.DOTALL|re.MULTILINE)
        links = re.findall("<a href=(.*?)>", post, re.DOTALL|re.MULTILINE)
        #title = re.sub('\<.*?\>', '', contents[0])
        message = re.sub('\<.*?\>', '', contents[1])
        page = requests.get('http://www.ppomppu.co.kr/'+links[0])
        title = re.findall("<!--DCM_TITLE-->(.*?)<!--/DCM_TITLE-->", page.text, re.DOTALL|re.MULTILINE)
        dates = re.findall("등록일: (.*?)<br>", page.text, re.DOTALL|re.MULTILINE)
        regdate = datetime.strptime(dates[0], '%Y-%m-%d %H:%M')
        if regdate > lastime:
            print(f'Title: {title}\nDate: {regdate}')#'\nMessage:{message}')
    except Exception as ex:
        pass

```
